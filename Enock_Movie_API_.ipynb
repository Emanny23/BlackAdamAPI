### Cell 3: Create a variable name that will store a search query for specific tweets. Then, create a different variable name to collect all of the different types of metadata contained in the search query. Last step: print out the data as text. 

Check out the data dictionaries on the Twitter Developer website to see what metadata from the tweet and user objects that you can collect via the API: https://developer.twitter.com/en/docs/twitter-api/data-dictionary/object-model/tweet 

### As you explore under cell 5, We can see the different tweet data dictionaries being collected. The first data dictionary that we see is tweet_fields, which is basically indicating the data about tweets and when we look even closer into its brackets after the equal sign we see three different 'values', the first is "created_at" which basically means the date and time the specific tweets in the data that you see were posted. The 2nd value is called text which displays the content of the tweets. The 3rd value is called source which is the name of the app used to post the tweet. In this case I think we should all know that we are only using twitter's data.

```python
query = "#BlackAdam -has:media -is:retweet"

tweets = client.search_recent_tweets(query = query,
                                     tweet_fields = ["created_at", "text", "source"],
                                     user_fields = ["name", "verified"],
                                     max_results = 100,
                                     expansions = ["author_id","attachments.media_keys"],
                                     media_fields = "duration_ms") 
print(tweets)
```

### Cell 7: Create an empty list to store the data, loop through each of the tweets in the data we collected above, and then extract specific kinds of tweet data

### What this data shows is the different compoments within a single tweet data collected, so basically lets say you tweeted about the Black Adam movie on twitter and said, something about the movie being too hyped. what you see in cell 7 basically will show the value labeled name which what ur username is, it'll idiciate whether you are verified, tweet text will show what you posted, source will basically display whom u are replying to, and finally the attachments will basically insicatge the tweet that you are retweeting. 

```python
tweet_list = []

for tweet, user in zip(tweets.data, tweets.includes["users"]):
    tweet_info = {
        "created_at": tweet.created_at,
        "name": user.name,
        "verified": user.verified,
        "text": tweet.text,
        "source": tweet.source,
        "attachments": tweet.data}
    tweet_list.append(tweet_info)
```
